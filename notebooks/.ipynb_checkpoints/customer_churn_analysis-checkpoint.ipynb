## Investigation of Missing TotalCharges Values

A detailed investigation revealed 11 records with blank values in the TotalCharges column.

Key findings:

- All affected customers have a tenure of 0 months.
- All affected customers are active (Churn = No).
- Since these customers have not completed any billing cycle, TotalCharges has not yet accumulated.

As the affected records represent only 11 out of 7,043 observations (0.16% of the dataset), they will be removed during the data-cleaning process to maintain data consistency.

```python
df['TotalCharges'] = df['TotalCharges'].replace(' ', np.nan)
```

```python
df['TotalCharges'].isnull().sum()
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])
```

```python
df.info()
```

# Data Cleaning Summary

### Cleaning Performed

- Investigated hidden missing values in the TotalCharges column.
- Identified 11 records containing blank values.
- Found that all affected customers had tenure = 0 months.
- Removed the 11 records as they represented only 0.16% of the dataset.
- Converted TotalCharges from object to float64.

### Result

The dataset now contains 7,032 clean records and is ready for exploratory data analysis.
